**Bài tập mẫu: Tính tổng các giá trị chẵn, tổng các giá trị lẻ, tổng các giá trị trong một dãy số.**

````
def sumEven(a):
    s = 0
    for e in range(0,len(a)):
        if(a[e] % 2 == 0):
            s = s + a[e]

    return s


def sumOdd(a):
    s = 0
    for e in a:
        if(e % 2 == 1):
            s = s + e

    return s


def sumList(a):
    return sum(a)


n = int(input('Nhập số phần tử trong dãy số n = '))

# aList = []
aList = [0 for j in range(0, n)]

for i in range(0, n):
    aList[i] = int(input('list[' + str(i) + '] = '))

print('Tổng các giá trị chẵn trong dãy = ', sumEven(aList))
print('Tổng các giá trị lẻ trong dãy = ', sumOdd(aList))
print('Tổng các giá trị trong dãy = ', sumList(aList))
````

Chương trình thực hiện nhập vào số nguyên n, khai báo một danh sách kích thước n phần tử với giá trị ban đầu bằng 0 (aList = [0 for j in range(0, n)])

````
def sumEven(a):
    return sum(a[i] for i in range(0, len(a)) if a[i]%2 == 0)


def sumOdd(a):
    return sum(i for i in a if i%2 == 1)


def sumList(a):
    return sum(a)


n = int(input('Nhập số phần tử trong dãy số n = '))

aList = [] # aList = list()

for i in range(0, n):
    aList.append( int(input('list[' + str(i) + '] = ')))
print('Tổng các giá trị chẵn trong dãy = ', sumEven(aList))
print('Tổng các giá trị lẻ trong dãy = ', sumOdd(aList))
print('Tổng các giá trị trong dãy = ', sumList(aList))

print(aList.append.__doc__)

````

**Bài tập mẫu: Đếm số từ trong văn bản**

````
def wordCount(s):
    token = s.split(' ')
    w = []
    c = []
    for t in token:
        if t not in w:
            w.append(t)
            c.append(count(t,token))
    return w, c



def count(w, token):
    count = 0
    for t in token:
        if w == t:
            count +=1
    return count


text = "lúa nếp là lúa nếp làng lúa lên lớp lớp lòng nàng lâng lâng"

tokens = text.split(' ')

print('tokens = ',tokens)

words, counts = wordCount(text)

for i in range(0,len(words)):
    print(words[i],':',counts[i])

print(words)
print(counts)
````

````
def wordCount(s):
    tokens = s.split(' ')
    words = list(set(tokens))
    counts = [tokens.count(word) for word in words]
    return words, counts

text = "lúa nếp là lúa nếp làng lúa lên lớp lớp lòng nàng lâng lâng"

text = "lúa nếp là lúa nếp làng lúa lên lớp lớp lòng nàng lâng lâng"

tokens = text.split(' ')

print(tokens)

words = set(tokens)

print(words)

words, counts = wordCount(text)

for i in range(0,len(words)):
    print(words[i],':',counts[i])
````

**Bài tập mẫu: Nhập và in ma trận chuyển vị**

````
def inputMatrix(matrix):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[i])):
            matrix[i][j] = int(input('matrix['+str(i)+']['+str(j)+']='))

def printMatrix(matrix):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[i])):
            print(matrix[i][j],end=" ")
        print()

def transMatrix(matrix):
    n = len(matrix)
    m = len(matrix[0])
    t = [[0 for j in range(0,n)] for i in range(0,m)]

    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[i])):
            t[j][i] = matrix[i][j]

    return t

m = int(input('Nhập vào số hàng = '))
n = int(input('nhập vào số cột = '))
matrix = [[0 for j in range(0,n)] for i in range(0,m)]

inputMatrix(matrix)
printMatrix(matrix)
t = transMatrix(matrix)
print('Ma trận chuyển vị')
printMatrix(t)
````

# Code ví dụ cho bài 10, sinh ma phương bậc lẻ

````
def magicSquare(m):
    n = len(m)
    i = 0
    j = n//2
    t = 1

    while(t <= n*n):
        m[i][j] = t
        t +=1
        i = i - 1
        j = j + 1
        if i < 0 and j == n:
            i = i+2
            j = j-1
        elif i < 0:
            i = n-1
        elif j == n:
            j = 0
        if m[i][j] > 0:
            i = i+2
            j = j-1



def printMatrix(matrix):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[i])):
            print(matrix[i][j],end="\t")
        print()


n = int(input('Nhập n = '))
m = [[0 for j in range(0,n)] for i in range(0, n)]
magicSquare(m)
printMatrix(m)
````
