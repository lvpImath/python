Matplotlib là thư viện vẽ đồ thị rất hữu ích cho những người làm việc với Python và Numpy, thư viện này giúp việc trực quan hóa dữ liệu trở nên dễ dàng và hiệu quả.

Bên cạnh việc cung cấp các hàm vẽ biểu đồ, matplotlib cũng cung cấp các hàm để vẽ hình học, thao tác xử lý với ảnh,...


# Các thành phần chính trong một đối tượng đồ thị

* **Figure**: Khung hình chứa các đối tượng đồ họa khác sẽ được hiển thị trên đó, nó giống như một khung tranh.
* **Axes**: Thành phần chính của một figure là các axes (những khung nhỏ hơn để vẽ hình lên đó). Một figure có thể chứa một hoặc nhiều axes. Nói cách khác, figure chỉ là khung chứa, chính các axes mới thật sự là nơi các hình vẽ được vẽ lên.
* **Axis**: Các chiều của đồ thị
* **Artist**: Các đối tượng đồ họa tạo nên các Axes như Text objects, Line2D objects, collection objects. Hầu hết các Artists được gắn với Axes.

# Bắt đầu với matplotlib

import thư viện và vẽ một biểu đồ đơn giản biểu diễn dữ liệu

```
#dòng này chỉ cần dùng trên Google Colab
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

plt.plot([1,2,3,4,5],[5,4,3,2,1])
plt.show()
plt.plot([1,2,3,4,5],[10,8,6,8,10])
plt.show()
```

Tiếp đến chúng ta có thể đặt tên cho biểu đồ và tên cho các trục tọa độ thông qua các thuộc tính **title**, **xlabel**, **ylabel**.

```
x = np.arange(-10,11)
y = x*x + 4
print(x)
print(y)
plt.plot(x,y)
plt.title('Parapolic')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Kích thước của khung hình biểu diễn đồ thị có thể được thiết lập thông qua tham số figsize trong phương thức figure()

```
x = np.arange(-10,11)
y =  - x*x - 4
plt.figure(figsize=(4,5))
plt.plot(x,y)
plt.title('Parapolic')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Đồ thị có thể được biểu diễn thông qua đường, hoặc các điểm, với hình dạng và mầu sắc khác nhau thông qua chuỗi đánh dấu.
Chuỗi đánh dấu định dạng mặc định là **"b-"** tương ứng là đường màu xanh. Ví dụ dưới đây sử dụng 2 chuỗi định dạng là **"or"** tương ứng là hình tròn màu đỏ, và **"kD"** tương ứng là hình kim cương màu đen.

```
x = np.arange(-10,11)
y1 =  - x*x - 4
y2 = -y1-100

plt.figure(figsize=(10,5))
plt.plot(x,y1,'or',x,y2,'kD')
plt.title('Parapolic')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Để vẽ nhiều biểu đồ tách biệt trong một khung hình figure, ta sử dụng phương thức subplot(). Phương thức này nhận 3 đối số nrows, ncols và index để xác định số lượng hàng, số cột và vị trí của sub-plot.


```
x = np.arange(-10,11)
y1 =  - x*x - 4
y2 = -y1

plt.subplot(1,2,1)
plt.plot(x,y1,'or')

plt.title('Parapolic down face')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1,2,2)
plt.plot(x,y2,'kD')

plt.title('Parapolic up face')
plt.xlabel('x')
plt.ylabel('y')
plt.subplots_adjust(wspace=0.4)
plt.show()
```

Để thiết lập các thông số giãn cách giữa các đồ thị con ta sử dụng phương thức subplots_adjust()
Tương tự, nếu ta muốn chia thành 4 đồ thị con như sau:

```
x = np.arange(-10,11)
y1 =  - x*x - 4
y2 = -y1

plt.subplot(2,2,1)
plt.plot(x,y1,'or')

plt.title('Parapolic down face 1 ')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,2)
plt.plot(x,y2,'kD')

plt.title('Parapolic up face 1')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,3)
plt.plot(x,y1,'or')

plt.title('Parapolic down face 2 ')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,4)
plt.plot(x,y2,'kD')

plt.title('Parapolic up face 2')
plt.xlabel('x')
plt.ylabel('y')

plt.subplots_adjust(wspace= 0.3,hspace= 0.6)
plt.show()
```

Chúng ta cũng có thể tạo ra các đồ thị con thông qua phương thức subplots(). Phương thức này trả lại 2 đối tượng là figure và axes có thể truy cập theo chỉ số (như chỉ số mảng) để thiết lập các tham số của figure và axes

```

fig, ax = plt.subplots(nrows = 3, ncols= 2, figsize = (5,5))

typ = np.random.choice(['b^','kD','or','vm'], 6)


for i in range(0,3):
  for j in range(0,2):
    ax[i,j].plot(np.random.randint(0,high= 10, size=10),np.random.randint(0,high= 10, size=10),typ[i*2+j])
    ax[i,j].set_title('Chart '+str((i*2)+j))
plt.subplots_adjust(wspace= 0.5,hspace= 0.5)
plt.show()
```

# Các loại biểu đồ

matplotlib cung cấp một số loại biểu đồ khác nhau. Tùy và dữ liệu, mục đích thể hiện, người dùng chọn loại biểu đồ cho phù hợp.

## Biểu đồ cột
 * bar() cột dọc
 * barh() cột ngang

```
labels = ['A','B','C','D','E','F','G']
values = np.random.randint(5,20,size = 7)

plt.bar(labels,values, color = 'magenta')
plt.title('Bar Chart Sample')
plt.xlabel(' Nhãn ')
plt.ylabel(' Giá trị ')
plt.show()
plt.barh(labels,values, color = 'green')
plt.show()
```

## Biểu đồ cột chồng


```
labels = ['A','B','C','D','E','F','G']
values1 = np.random.randint(5,20,size = 7)
values2 = np.random.randint(5,20,size = 7)
values3 = np.random.randint(5,20,size = 7)
index = np.arange(0,14,2)

width = 0.5

plt.bar(index,values1,width=width, label = 'Loại 1', color = 'pink')
plt.bar(index+width,values2,width=width, label = 'Loại 2', color = 'green')
plt.bar(index+2*width,values3,width=width, label = 'Loại 3', color = 'blue')

plt.xticks(index+width,labels)
plt.title('Biểu đồ cột chồng theo chiều ngang')
plt.xlabel(' Nhãn ')
plt.ylabel(' Giá trị ')
plt.legend(loc = 'best')

plt.show()
```

```
labels = ['A','B','C','D','E','F','G']
values1 = np.random.randint(5,20,size = 7)
values2 = np.random.randint(5,20,size = 7)
values3 = np.random.randint(5,20,size = 7)
index = np.arange(0,14,2)

width = 0.5

plt.bar(index,values1,width=width, label = 'Loại 1', color = 'pink')
plt.bar(index,values2,width=width, label = 'Loại 2', color = 'green', bottom=values1)
plt.bar(index,values3,width=width, label = 'Loại 3', color = 'blue',bottom=values1+values2)

plt.xticks(index,labels)
plt.title('Biểu đồ cột chồng theo chiều dọc')
plt.xlabel(' Nhãn ')
plt.ylabel(' Giá trị ')
plt.legend(loc = 'best')

plt.show()
```

## Biểu đồ tròn


```
labels = ['A','B','C','D','E','F','G']
values = np.random.randint(5,20,size = 7)

expl = [0,0.1,0,0,0,0.2,0]

plt.pie(values,explode=expl,labels = labels, shadow= True,startangle=90, autopct='%1.1f%%')

plt.show()
```

## Biểu đồ Histogram

Histogram là một loại biểu đồ rất phổ biến khi chúng ta xem xét dữ liệu như chiều cao và cân nặng, giá cổ phiếu, thời gian chờ đợi của một khách hàng, v.v ... liên tục trong tự nhiên. Histogram's data được vẽ trong một phạm vi so với tần số của nó. Histograms là các biểu đồ xuất hiện rất phổ biến trong xác suất và thống kê và tạo cơ sở cho các distributions khác nhau như normal -distribution, t-distribution, v.v.


```
data =  np.random.randint(-5,5,size=100000)

plt.title('Histogram')
plt.xlabel('Giá trị')
plt.ylabel(' Số lần ')
plt.hist(data, 10, color='green')
plt.show()

data =  np.random.randn(1000000)

plt.title('Histogram')
plt.xlabel('Giá trị')
plt.ylabel(' Số lần ')
plt.hist(data, 100, color='blue')
plt.show()

```

## Biểu đồ phân tán

Các biểu đồ phân tán là các biểu đồ được sử dụng rộng rãi, đặc biệt là chúng có ích trong việc hình dung một vấn đề về hồi quy. Trong ví dụ sau, cung cấp dữ liệu được tạo tùy ý về chiều cao và cân nặng và vẽ chúng với nhau. Mình đã sử dụng các phương thức xlim() và ylim() để đặt giới hạn của trục X và trục Y tương ứng.



```
h = np.random.randint(140,200,size=20)
w = np.random.randint(60,100,size=20)

plt.xlim(140,200)
plt.ylim(60,100)
plt.scatter(h,w)
plt.title(' Biểu đồ phân tán')
plt.xlabel(' Chiều cao ')
plt.ylabel(' Cân nặng ')
plt.show()
```
