````
# Một lớp mô phỏng một hình chữ nhật.
class Rectangle :
  '''
  This is Rectangle class
  '''

  # Hàm dựng (Contructor).
  def __init__(self, width, height):
    self.width = width
    self.height = height

  def getWidth(self):
    return self.width

  def getHeight(self):
    return self.height

  # Phương thức tính diện tích.
  def getArea(self):
    return self.width * self.height
    # return getWith()*getHeight()
    # return self.getWith() * self.getHeight()

````

## TÍNH ĐÓNG GÓI - ENCAPSULATION (8)
* Các thuộc tính và phương thức của lớp trong Python mặc định là không hạn chế truy cập (pubic), chúng ta có thể truy cập tới mọi thành phần của lớp từ môi trường bên ngoài của lớp.
* Chúng ta có thể hạn chế quyền truy cập vào trạng thái bên trong của đối tượng. Điều này ngăn chặn dữ liệu bị sửa đổi trực tiếp, được gọi là đóng gói.
* Trong Python, chúng ta biểu thị thuộc tính private này bằng cách sử dụng dấu gạch dưới làm tiền tố: “_” (protected) hoặc “__“ (private).

````
# Ví dụ về lớp với các thuộc tính public

class Student:
  schoolName = 'HUS School' # class attribute

  def __init__(self, name, age):
      self.name=name # instance attribute
      self.age=age # instance attribute

  def printInfor(self):
    print(self.name, self.age, self.schoolName)


# Các thuộc tính của lớp có thể được truy cập và
# thay đổi giá trị của chúng từ môi trường bên ngoài

std = Student("MIM", 65)
std.printInfor()

print("Change age:")
std.age = 20
std.printInfor()
print("Change schoolName of class")
Student.schoolName = "HN VNU HUS"
std.printInfor()
print("Change schoolName of object")
std.schoolName = "VNU HUS"
std.printInfor()

print("Class schoolName:",Student.schoolName)


newstd = Student("MIM-HUS",66)
newstd.printInfor()
````

## Các thành phần hạn chế quyền truy cập (protected)
* Các thành phần bị hạn chế quyền truy cập trong Python được xác định bởi các định danh bắt đầu bằng dấu gạch dưới _
* Với việc hạn chế quyền truy cập, một phần tử protected chỉ có thể được truy cập đến từ chính lớp đó hoặc từ các lớp con của nó.


````
class Student():
  _schoolName = 'HUS School' # protected class attribute

  def __init__(self, name, age):
      self._name=name  # protected instance attribute
      self._age=age # protected instance attribute

  def printInfor(self):
    print(self._name, self._age, self._schoolName)


std = Student("MIM", 65)
std.printInfor()

std._name = "GEO"
print("Change name to:",std._name)

std.printInfor()

Student._schoolName = "VNU HUS"
print("Change class schoolName:", Student._schoolName)
std.printInfor()

````

### KẾ THỪA - INHERITANCE (9)

* Tính kế thừa cho phép một lớp (class) có thể kế thừa các thuộc tính và phương thức từ các lớp khác đã được định nghĩa.
* Lớp con kế thừa tất cả thành phần của lớp cha, có thể mở rộng các thành phần kế thừa và bổ sung thêm các thành phần mới


````
# Lớp cha
class Car:       # mặc định (object)
  # Constructor
  def __init__(self, brand, name, color):
    self.brand = brand
    self.name = name
    self.color = color
  # phương thức
  def run(self):
    print ("{} đang chạy trên đường".format(self.name))
  def stop(self, task):
    print ("{} đang dừng xe để {}".format(self.name, task))


# Lớp Toyota mở rộng từ lớp Car.
class Toyota(Car):
  def __init__(self, brand, name, color, fuel):
    # Gọi tới constructor của lớp cha (Car)
    # để gán giá trị vào thuộc tính của lớp cha.
    super().__init__(brand, name, color)
    self.fuel = fuel

  # Ghi đè (override) phương thức cùng tên của lớp cha.
  def stop(self, task):
    print ("{} đang dừng xe để {}".format(self.name, task))
    print ("{} chạy bằng {}".format(self.name, self.fuel))

  # Bổ sung thêm thành phần mới
  def start(self):
    print ("{} đang nổ máy".format(self.name))


toyota1 = Toyota("Toyota", "Toyota Hilux", "Đỏ", "Điện")
toyota2 = Toyota("Toyota", "Toyota Yaris", "Vàng", "Deisel")
toyota3 = Toyota("Toyota", "Toyota Vios", "Xanh", "Gas")
toyota1.stop("nạp điện")
toyota2.run()
toyota3.start()

````

### TÍNH ĐA HÌNH - POLYMORPHISM (10)

* Tính đa hình là khái niệm mà hai hoặc nhiều lớp có những phương thức giống nhau nhưng có thể thực thi theo những cách thức khác nhau.

* Về cơ bản, tính đa hình trong Python thể hiện rất rõ trong cơ chế kiểu dữ liệu động.

````
class Toyota:

  def stop(self):
    print("Toyota dừng xe để nạp điện")

  def start(self):
    print("Toyota nổ máy bằng hộp số tự động")

class Porsche:

  def stop(self):
    print("Porsche dừng xe để bơm xăng")

  def start(self):
    print("Porsche nổ máy bằng hộp số cơ")

# common interface
def carStop(car):
  car.stop()
# instantiate objects
toyota = Toyota()
porsche = Porsche()
# passing the object
carStop(toyota)
carStop(porsche)
````
