## Kiến thức

Sinh viên được hướng dẫn các kiến thức cơ bản và nâng cao về ngôn ngữ lập trình Python như:
* Các khái niệm cơ bản về ngôn ngữ lập trình Python (biến, kiểu dữ liệu, không gian tên, phép gán, tham chiếu,...)

* Cách sử dụng cấu trúc điều khiển, vòng lặp,...

* Các kiểu dữ liệu cơ bản trong ngôn ngữ Python (Danh sách, Tập hợp, Từ điển,...)
* Làm việc với tệp, xử lỗi và ngoại lệ,...
* Lập trình hướng đối tượng, thiết kế và tạo lớp (class) với Python

* Một số thư viện rất hữu dụng trong việc tính toán như numpy, pandas, sklearn, matplotlib,...


## Kỹ năng

* Sinh viên làm quen và lập trình theo các phong cách của ngôn ngữ Python
* Sử dụng ngôn  ngữ Python để viết các chương trình giải quyết các bài toán được đặt ra.
* Từ những kiến thức cơ bản được trang bị trên lớp, sinh viên cần tự tìm hiểu và học cách sử dụng các thư viện khác của Python để xây dựng các ứng dụng cụ thể mà sinh viên có hứng thú hoặc được giáo viên yêu cầu như:
  * Sử dụng Python để xử lý dữ liệu, trực quan hóa dữ liệu
  * Lập trình giải các bài toán về Học máy như xử lý ảnh, xử lý văn bản...
  * Lập trình ứng dụng, game, web...









## Thái độ

* Sinh viên cần có thái độ học tập tích cực, chủ động, tránh tâm lý tiêu cực như chủ quan với những bài đầu, khi gặp vấn đề khó lại chán nản không chủ động hỏi giáo viên hoặc các bạn cùng lớp.
* Chủ động trong việc trao đổi với giáo viên về các vấn đề gặp phải, bên cạnh đó cũng rất cần có tinh thần tự học và tìm hiểu các tài liệu có trên internet.
* Cần theo sát chương trình học trên lớp, hoàn thành các bài tập được giao
* Cần có ý thức tự giác cao, không copy bài giải của các bạn và cũng không cho các bạn khác copy bài tập của mình
* Có tinh thần hợp tác và trách nhiệm trong việc hoàn thành nhiệm vụ học tập của mình với môn học cũng như khi làm bài tập lớn với nhóm.
