## Xây dựng hàm

* Một chương trình sẽ dễ dàng tiếp cận, dễ đọc, dễ hiểu và dễ kiểm soát lỗi nếu được tổ chức thành các hàm chức năng một cách hợp lý.
* Việc viết các hàm thực hiện các công việc thành phần trong một chương trình giúp tái sử dụng hàm đó với cùng một công việc được thực hiện nhiều lần, dễ khoanh vùng xác định lỗi và sửa lỗi.

* Mỗi hàm trong Python có hai khía cạnh

>* Định nghĩa hàm **(Function definition)**: định nghĩa của một hàm chứa các dòng lệnh xác định các hành vi của hàm đó.

>* Lời gọi hàm **(Function invocation)**: Một hàm được sử dụng trong chương trình thông qua lời gọi hàm


* Mỗi hàm được định nghĩa một lần nhưng có thể gọi đến nhiều lần.
* Khuôn mẫu để định nghĩa một hàm như sau:

illus25.png


```
def get_average(p1, p2, p3):
  avg = (p1+p2+p3)/3
  return avg

print(get_average(1,2,3))
```

## Các thành phần trong hàm

* Trong ví dụ trên, từ khóa def đánh dấu việc bắt đầu định nghĩa hàm, tên hàm là `get_average`, đối số là `p1, p2, p3`, nội dung hàm là trả lại giá trị trung bình của 3 đối số nhập vào thông qua phép gán `avg = (p1+p2+p3)/3`
*  Lời gọi hàm được thực hiện bằng tên hàm (`get_average`)
cùng với đối số được truyền vào là `1, 2, 3`, kết quả lời gọi hàm `get_average(1, 2, 3)` được in ra màn hình bằng lệnh `print`.

* Từ khóa `return` được dùng để trả lại giá trị cho hàm.
Một hàm có thể không có đối số, và có thể không
cần trả lại giá trị gì

```
def print_info(person):
  print("Họ tên: %s, Giới tính: %s, năm sinh: %d" %(person[0],person[1],person[2]))

person = ["Toán Văn Anh", "Nam", 2001]

print_info(person)
```

## Đối số của hàm

* Khi một hàm được gọi tới với tham số được
truyền vào, khi đó biến tham số sẽ tham chiếu
đến biến hoặc giá trị được truyền làm tham số
*  Các kiểu dữ liệu như integer, float-point và
string được xem là các đối tượng bất biến
(immutable), giá trị của các đối tượng bất biến
không bị thay đổi khi được truyền là tham số
của các hàm
* Trong ví dụ dưới đây, kết quả của double(x) là 6,
nhưng x vẫn có giá trị bằng 3


```
def double(a):
  a = a*2
  return a

x = 3
print(double(x))
print(x)
```

```
def double_list(a):
  for i in range(len(a)):
    a[i] = a[i]*2
  # return a

x = [1, 2, 3, 4]
print(double_list(x))
print(x)
```

## Tham số mặc định

* Tham số của hàm có thể có giá trị mặc định, giá trị này có thể bỏ qua trong lời gọi hàm, các tham số có giá trị mặc định phải đứng sau các tham số không có giá trị mặc định.
* Trong ví dụ dưới đây, tham số a của hàm grow có
giá trị mặc định là 2, khi gọi hàm grow ta có thể bỏ qua tham số a, khi đó a sẽ nhận giá trị mặc định là
2, ta cũng có thể thay giá trị mặc định của a bằng
giá trị khác.

```
def grow(b, a=2):
  return a*b

x = 5
print(grow(x))
print(grow(x,4))
print(grow('abc',3))
```

## Một số ví dụ về hàm với tham số có tiền tố `*` và `**`

* Python sử dụng tiền tố `*` để đánh dấu một tham số là không cố định về số lượng tham số
* Nếu tham số được đánh dấu là `**` thì mỗi đối số là 1 cặp khóa - giá trị

```
def multiple(*x):
  result = 1
  for i in x:
    result *=i # result = result*i

  return result

print(multiple())
print(multiple(1,2,3,4))
print(multiple(2,3))
print(multiple(1,3.5,6,7))
print(multiple([1,2,3,4]))
print(multiple(*[1,2,3,4]))
print(*[1,2,3,4,5])
```

## Thứ tự các đối số
Khi xây dựng hàm, cần chú ý đến thứ tự các đối số
1. Đối số không có giá trị mặc định
2. Đối số có giá trị mặc định
3. Đối số dạng `*args` (danh sách)
3. Đối số dạng `**kwargs` (dạng từ điển, khóa-giá trị)

```
def args_order(basic_args, basic_default_args=2, *args, **kwargs):
  salary = basic_args * basic_default_args
  bonus = 0
  for arg in args:
    if arg in kwargs:
      bonus += kwargs[arg]

  return salary + bonus

boss_salary = args_order(100, 3,'a','b',a=10,b=15,c=17,d=20)
print('boss:',boss_salary)

bonus_dict = {'a':10,'b':15,'c':17,'d':20}

dev_salary = args_order(80)
print('dev:', dev_salary)

# Chú ý trong trường hợp này biến mặc định basic_default_args được xác định theo vị trí

dev_salary = args_order(80,5,*['c','d'],**bonus_dict) #có lỗi
print('dev:', dev_salary)
```

## Phạm vi biến

* Biến được định nghĩa bên trong hàm được gọi
là biến cục bộ (local variable). Biến cục bộ được
lưu trong bộ nhớ khi hàm được gọi đến và thực
thi, khi kết thúc hàm hoặc biến không còn được
sử dụng tới trong hàm thì bộ nhớ dùng để lưu
trữ biến sẽ được giải phóng. Cùng một tên biến
có thể sử dụng trong các hàm khác nhau mà
không gây xung đột.

* Trái ngược với biến cục bộ là biến toàn cục
(global variable), biến toàn cục có phạm vi trên
toàn bộ chương trình, trên tất cả các hàm, điều
này có nghĩa là tất cả các hàm đều có thể truy
cập và thay đổi biến toàn cục. Để khai báo sử
dụng biến toàn cục ta sử dụng từ khóa global,
việc sử dụng từ khóa global chỉ cần thiết khi ta
gán, hoặc thay đổi giá trị của biến toàn cục.


```
# result = 1
x, y, z = 5, 6, 3
rs = 1
def maxInTriple():
  local_var = max(max(x,y), z) #
  return local_var

def doubleMax():
  global rs
  rs = maxInTriple() * 2
  print('local result',rs)
  #print('global result',result)

print(maxInTriple())
doubleMax()
print(rs) ##  1 vs 12 ????

# print(local_var)
```

* Các quy tắc về phạm vi của biến sẽ được nêu rõ hơn trong bài học về Phạm vi và không gian tên (scope and namespace)


```
a_global_var = 9
a =101
def bounder_function():
  a_non_local_var = 20

  def inner_function():

    x = a
    a_local_var = 100
    print('inner print')
    a_global_var = -1
    a_non_local_var = -2
    print(a_global_var, a_non_local_var, a_local_var)

  inner_function()
  print('bounder print')
  # print(a_global_var, a_non_local_var, a_local_var)
  # global a_global_var
  a_global_var = -99
  print(a_global_var, a_non_local_var)


bounder_function()

print('global print')
print(a_global_var)
```

## Hàm đệ quy

* Hàm đệ quy là các hàm có chứa lời gọi hàm đến
chính nó. Hàm đệ quy thường được sử dụng để
giải các bài toán có định nghĩa đệ quy hoặc quy
nạp
* Nhìn chung, với cùng một bài toán nếu có hàm
đệ quy và hàm không đệ quy để giải bài toán
đó thì thường hàm đệ quy có lời giải ngắn gọn
hơn, tuy nhiên trong một số trường hợp hàm đệ
quy sẽ tốn bộ nhớ và chi phí tính toán hơn.

```
def factorial(n):
  if n == 0:
    return 1
  else:
    return n* factorial(n-1)

print(factorial(-6))
```

```
a = [1, 2, 3, 4, 1, 1, 1,1,1,1, 2, 2, 2, 1, 3, 4, 4, 4 ]
i = 0

while i < len(a)-1:
  if a[i] == a[i+1]:
    del a[i]
    i -= 1

  i +=1

print(a)

```

## Lệnh `break`
* Các vòng lặp có thể kết
thúc bằng lệnh `break`
* Lệnh `break` thường
được dùng để kết thúc
vòng lặp khi đạt được
điều kiện không, hoặc
khó xác định bởi số lần
lặp, đôi khi là vòng lặp
vô hạn cho đến khi tìm
thỏa mãn điều kiện thì
dừng

image.png

```
import math
a , b = 1, 10 # hai điểm đầu mút của đoạn [1, 10]
e = 0.000000001 # sai số
f = lambda x : x**3-8
y = (a+b)/2
looping = True
while looping and math.fabs(f(y)) > e:
  y = (a+b)/2
  if math.fabs(f(y)) <= e:
    looping = False
  elif f(y)*f(a) < 0:
    b = y
  else:
    a = y
print('nghiệm của phương trình là y = ',y)
```
