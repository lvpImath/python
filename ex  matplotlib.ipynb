## Hướng dẫn bài 1 

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'
data  =  np.genfromtxt(url, delimiter=',',dtype=None, encoding=None)
print('Dữ liệu đọc được từ file csv\n',data)
print(data.shape)
labels = data[0,:]

print('Các nhãn dữ liệu\n',labels)
# Loại bỏ dòng nhãn
labels = np.delete(labels,[0,0])
data = np.delete(data, 0, 0)
months =  data[:,0:1]
print('Các tháng \n',months)
data = np.delete(data,0,1)

data =  data.astype(int)
print('Dữ liệu sau khi bỏ nhãn và cột tháng',data)
print('kích thước dữ liệu\n',data.shape)

months = np.reshape(months,(12))

profit = np.reshape(data[:,7:8],(12))

print(profit, months)
plt.plot(months,profit)
plt.title('Biểu đồ lợi nhuận theo từng tháng')
plt.xlabel(' Tháng ')
plt.ylabel(' Lợi nhuận ')

```

# Bài 3.

Hiển thị biểu đồ dạng đường, các sản phản bán được theo từng tháng, mỗi loại sản phẩm trên 1 đường.

# Bài 4.

Hiển thị biển đồ phân tán (scatter) của dữ liệu kem đánh răng (toothpaste) và dầu gội (shampoo) trên 2 biểu đồ con khác nhau



```
# Bài 4
```

## Hướng dẫn bài 4

```

plt.figure(figsize=(20,5))
plt.subplot(1,2,1)

plt.scatter(months, data[:,[2]], label = 'Kem đánh răng')
plt.legend(loc='upper left')
plt.title('Kem đánh răng')
plt.xlabel(' Tháng ')
plt.ylabel(' Số sản phẩm bán ra ')
plt.yticks([i for i in range(4500, 10000,500)])
plt.grid(True, linewidth= 1, linestyle="--")


plt.subplot(1,2,2)

plt.scatter(months, data[:,[4]], label = 'Dầu gội', color = 'r')
plt.legend(loc='upper left')
plt.title('Dầu gội')
plt.xlabel(' Tháng ')
plt.ylabel(' Số sản phẩm bán ra ')
plt.yticks([i for i in range(1500, 5000,500)])
plt.subplots_adjust(wspace=0.5)
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```

Đọc dữ liệu về số sản phẩm bán từng tháng của tất cả các loại sản phẩm, vẽ đồ thị vùng chồng.

```
# Bài 8
```

## Hướng dẫn bài 8


```
saledata = np.transpose(data[:,0:6])

print(saledata)
lbs =  labels[0:6]

plt.figure(figsize=(10,10))
plt.stackplot(months,saledata, labels = lbs, colors=['m','c','r','k','g','y'])


plt.xlabel('Tháng')
plt.ylabel('Số sản phẩm bán ra')
plt.title('Doanh số các sản phảm theo tháng')
plt.legend(loc='upper left')
plt.show()
```
