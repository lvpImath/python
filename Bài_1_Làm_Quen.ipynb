## Một số đặc điểm của ngôn ngữ Python
* **Học và sử dụng một cách dễ dàng**: Python dễ học và dễ sử dụng bởi vì nó là ngỗn ngữ lập trình bậc cao, dễ được tiếp cận bởi người dùng.
* **Ngôn ngữ biểu đạt (Expressive Language)**: Ngôn ngữ Python có tính biểu đạt cao có nghĩa là nó dễ đọc và dễ hiểu.
* **Ngôn ngữ thông dịch (Interpreted Language)**: Python là một ngôn ngữ thông dịch, trình thông dịch thực thi từng lệnh tại mỗi thời điểm, điều này giúp việc kiểm soát và gỡ lỗi dễ dàng hơn, do đó nó phù hợp với những người mới học lập trình.

* **Ngôn ngữ đa nền tảng (Cross-platform Language)**: Python hoạt động đồng nhất trên các nền tảng khác nhau như as Windows, Linux, Unix và Macintosh,… Do vậy chúng ta cũng có thể gọi Python là một ngôn ngữ portable (portable language).
* **Miễn phí và mã nguồn mở**: Giấy phép Python Software Foundation, Python được cung cấp miễn phí trên trang chủ https://www.python.org.
* **Ngôn ngữ hướng đối tượng (Object-Oriented Language)**: Python hỗ trợ ngôn ngữ lập trình hướng đối tượng và đưa ra các khái niệm về các lớp (class) và đối tượng (object).


## Một số đặc điểm của ngôn ngữ Python (tiếp)


* **Có khả năng mở rộng (Extensible)**: Python có thể viết mã chương trình bằng các ngôn ngữ khác như C, C++,… và cũng có thể biên dịch mã chương trình đó trong các ngôn ngữ tương ứng.
* **Thư viện khổng lồ (Large Standard Library)**: Python có một hệ thống thư viện khổng lồ và trải rộng trên các lĩnh vực, nó cung cấp một nguồn tài nguyên phong phú giúp việc phát triển các ứng dụng trở nên nhanh chóng.
* **Hỗ trợ lập trình giao diện (GUI Programming Support)**: Giao diện người dùng có thể được phát triển bằng Python.

* **Khả năng tích hợp (Integrated)**: Python có thể dễ dàng tích hợp với các ngôn ngữ khác như C, C++, JAVA,...
*  **Ngôn ngữ kiểu dữ liệu động (Dynamically Typed Language)**: Kiểu dữ liệu của một biến được quyết định khi thực thi chứ không phải khi khai báo, khởi tạo. Do vậy không cần khai báo kiểu dữ liệu cho các biến trong Python.

## Các thư viện cần thiết trong Python

Một trong nhưng ưu thế đáng kể của Python là cộng đồng rộng lớn, cùng với kho thư viện khổng lồ, đặc biệt là các thư viện liên quan đến tính toán khoa học và xử lý dữ liệu cũng như các thư viện về mô hình cũng như học máy.



### Numpy

NumPy, viết tắt của Numerical Python, từ lâu đã là nền tảng của tính toán số trong Python. Nó cung cấp cấu trúc dữ liệu, thuật toán và  thư viện keo cần thiết cho hầu hết các ứng dụng khoa học liên quan đến dữ liệu số trong Python. NumPy bao gồm:

* Ndarray đối tượng mảng đa chiều với khả năng truy cập nhanh chóng và hiệu quả trong việc tính toán

* Các hàm để thực hiện các phép tính thông minh giữa phần tử với các mảng hoặc các phép toán giữa các mảng với nhau

* Công cụ để đọc và ghi tập dữ liệu dựa trên mảng vào đĩa

* Các phép toán đại số tuyến tính, biến đổi Fourier và tạo số ngẫu nhiên

* Một API C hoàn thiện cho phép các phần mở rộng Python và mã C hoặc C++ gốc có thể truy cập các cấu trúc dữ liệu và cơ sở tính toán của NumPy

Ngoài khả năng xử lý mảng nhanh mà NumPy bổ sung vào Python, một trong những ứng dụng chính của nó trong phân tích dữ liệu là khả năng chứa dữ liệu được dùng để truyền giữa các thuật toán và thư viện. Đối với dữ liệu số, mảng NumPy hiệu quả hơn để lưu trữ và thao tác dữ liệu so với các cấu trúc dữ liệu Python tích hợp sẵn khác. Ngoài ra, các thư viện được viết bằng ngôn ngữ cấp thấp hơn, chẳng hạn như C hoặc FORTRAN, có thể hoạt động trên dữ liệu được lưu trữ trong mảng NumPy mà không cần sao chép dữ liệu vào một số biểu diễn bộ nhớ khác. Do đó, nhiều công cụ tính toán số cho Python sử dụng NumPy làm cấu trúc dữ liệu chính, hoặc được xây dựng với mục tiêu có thể tương tác với NumPy

### Pandas

pandas cung cấp các cấu trúc và chức năng dữ liệu cấp cao được thiết kế để làm việc với dữ liệu có cấu trúc hoặc dạng bảng một cách trực quan và linh hoạt. Kể từ khi xuất hiện vào năm 2010, nó đã giúp Python trở thành một môi trường phân tích dữ liệu hiệu quả và mạnh mẽ. Các đối tượng chính trong pandas là:
* DataFrame: một cấu trúc dữ liệu hướng cột, dạng bảng với cả nhãn hàng và cột
* Series một đối tượng mảng một chiều có nhãn.

Pandas pha trộn ý tưởng tính toán mảng của NumPy với các thao tác dữ liệu trong bảng tính và cơ sở dữ liệu quan hệ (chẳng hạn như SQL). Nó cung cấp chức năng lập chỉ mục thuận tiện để cho phép bạn định hình lại, cắt lát và xắt nhỏ, thực hiện tổng hợp và chọn tập hợp dữ liệu con. Vì thao tác chuẩn bị và làm sạch dữ liệu là những kỹ năng quan trọng trong phân tích dữ liệu

Pandas bắt đầu xây dựng vào đầu năm 2008  tại AQR Capital Management, một công ty quản lý đầu tư định lượng. Mục tiêu của pandas là:

* Cấu trúc dữ liệu với các trục được gắn nhãn hỗ trợ căn chỉnh dữ liệu tự động hoặc rõ ràng — điều này ngăn ngừa các lỗi phổ biến do dữ liệu bị lệch và làm việc với dữ liệu được lập chỉ mục khác nhau đến từ các nguồn khác nhau

* Tích hợp chức năng chuỗi thời gian

* Xử lý dữ liệu theo cấu trúc giống nhau bao gồm cả dữ liệu chuỗi thời gian và dữ liệu chuỗi không thời gian

* Các phép toán số học và rút gọn để bảo toàn siêu dữ liệu

* Xử lý linh hoạt các dữ liệu bị thiếu

* Hợp nhất và các phép toán quan hệ khác được tìm thấy trong cơ sở dữ liệu phổ biến (ví dụ: dựa trên SQL)

Bản thân tên pandas có nguồn gốc từ dữ liệu bảng, một thuật ngữ kinh tế lượng cho các tập dữ liệu có cấu trúc đa chiều và một cách chơi chữ của cụm từ phân tích dữ liệu Python (Python data analysis)

### matplotlib

matplotlib là thư viện Python phổ biến nhất để tạo ra các biểu đồ và các hình ảnh hóa dữ liệu hai chiều. Ban đầu nó được tạo ra bởi John D. Hunter và hiện được duy trì bởi một nhóm lớn các nhà phát triển. Nó được thiết kế để tạo ra các biểu đồ phù hợp cho việc xuất bản. Trong khi có các thư viện trực quan hóa khác có sẵn cho các lập trình viên Python, matplotlib vẫn được sử dụng rộng rãi và tích hợp hợp lý với phần còn lại của hệ sinh thái. matplotlib là một lựa chọn đơn giản và hiệu quả cho một công cụ trực quan hóa mặc định.

### SciPy

SciPy là một tập hợp các gói giải quyết một số vấn đề cơ bản trong khoa học tính toán. Dưới đây là một số công cụ mà nó chứa trong các mô-đun khác nhau của nó:

**scipy.integrate**

Quy trình tích phân số và trình giải phương trình vi phân

**scipy.linalg**

Các quy trình đại số tuyến tính và phân rã ma trận mở rộng bên ngoài những quy trình được cung cấp trong numpy.linalg

**scipy.optimize**

Bộ tối ưu hóa chức năng (bộ thu nhỏ) và thuật toán tìm gốc

**scipy.signal**

Các công cụ xử lý tín hiệu

**scipy.sparse**

Ma trận thưa và bộ giải hệ thống tuyến tính thưa

**scipy.special**

Bao quanh SPECFUN, một thư viện FORTRAN triển khai nhiều hàm toán học phổ biến, chẳng hạn như hàm gamma

**scipy.stats**

Phân phối xác suất liên tục và rời rạc tiêu chuẩn (hàm mật độ, bộ lấy mẫu, hàm phân phối liên tục), các thử nghiệm thống kê khác nhau và các thống kê mô tả khác


Cùng với nhau, NumPy và SciPy tạo thành một nền tảng tính toán hoàn chỉnh và hoàn thiện hợp lý cho nhiều ứng dụng tính toán khoa học truyền thống.

### scikit-learn

Dự án bắt đầu vào năm 2007, scikit-learning đã trở thành bộ công cụ học máy có mục đích chung hàng đầu dành cho các lập trình viên Python. Tính đến năm 2022, hơn 2.000 cá nhân khác nhau đã đóng góp mã cho dự án. Nó bao gồm các mô-đun con cho các mô hình như:

* Phân loại: SVM, hàng xóm gần nhất, rừng ngẫu nhiên, hồi quy hậu cần, v.v.

* Hồi quy: Lasso, hồi quy ridge, v.v.

* Phân cụm: k-mean, phân cụm spectral, v.v.

Giảm kích thước: PCA, lựa chọn đặc trưng, phân tích nhân tử ma trận, v.v.

Lựa chọn mô hình: Tìm kiếm theo lưới, cross-validation, metrics

Tiền xử lý: Trích chọn đặc trưng, chuẩn hóa

Cùng với pandas, mô hình thống kê và IPython, scikit-learning đã đóng vai trò quan trọng trong việc cho phép Python trở thành một ngôn ngữ lập trình khoa học dữ liệu hiệu quả.

### statsmodels
So với scikit-learning, mô hình thống kê chứa các thuật toán cho thống kê và kinh tế lượng (primarily frequentist) cổ điển. Thư viện này bao gồm các mô-đun con như:

Các mô hình hồi quy: hồi quy tuyến tính, mô hình tuyến tính tổng quát, mô hình tuyến tính mạnh mẽ, mô hình hiệu ứng hỗn hợp tuyến tính, v.v.

Phân tích phương sai (ANOVA)

Phân tích chuỗi thời gian: AR, ARMA, ARIMA, VAR và các mô hình khác

Phương pháp phi tham số: Ước tính mật độ nhân, hồi quy nhân

Trực quan hóa các kết quả mô hình thống kê

mô hình thống kê tập trung hơn vào suy luận thống kê, cung cấp các ước lượng không chắc chắn theo giá trị các tham số p. Ngược lại, scikit-learning tập trung vào hơn vào dự đoán.

# Một số ví dụ


```python
import sys
import time

chars = ['/','-','\\']
for i in range(100):
  sys.stdout.write(chars[i%3])
  sys.stdout.flush()
  time.sleep(0.2)
  sys.stdout.write('\r')
```

```python
import random

print('MIM Lucky Number')

number = random.randint(0,100)

print('Nhập vào 1 số từ 0 đến 99')

while True:
  guess = input()
  if guess:
    guess = int(guess)
    if guess == number:
      print('Chúc mừng bạn đã đoán chính xác con số may mắn :)')
      break
    elif guess > number:
      print('Bạn đoán chưa đúng số may mắn (*~*), hãy chọn số khác nhỏ hơn: ')
    else:
      print('Bạn đoán chưa đúng số may mắn (*0*), hãy chọn số khác lớn hơn: ')
  else:
    print('Chúc bạn may mắn lần sau :)')
    break
```
