## Bài 13. Lấy các giá trị trong mảng theo điều kiện.

Cho mảng a, lấy ra các giá trị x trong a thỏa mãn 5 	&le; x 	&le; 10

```
a = np.array([2, 6, 1, 9, 10, 3, 27])
```

````
# Bài 13
````

### Hướng dẫn bài 13

````
a = np.arange(15)

# Method 1
index = np.where((a >= 5) & (a <= 10))
print('Cách 1\n',a[index])

# Method 2:
index = np.where(np.logical_and(a>=5, a<=10))
print('Cách 2\n',a[index])

# Method 3:
print('Cách 3\n',a[(a >= 5) & (a <= 10)])
````

## Bài 18. Một số thao tác với định dạng in

Tạo một mảng ngẫu nhiên với kích thước 5x3

* Định dạng chỉ in ra 3 chữ số phần thập phân

* Chuyển từ định dạng mũ  e<sup>x</sup> sang dạng bình thường. Ví dụ 5.0e-02 => 0.05
* Cho mảng 1 chiều gồm 100 phần tử từ 1 đến 100, định dạng để in ra tối đa 6 phần tử trong mảng. Ngược lại, định dạng để in ra tất cả các phần tử trong mảng


````
# Bài 18
````

### Hướng dẫn bài 18

````
import numpy as np
# Tạo mảng các phần tử được sinh ngẫu nhiên, kích thước 5x3
rand_arr = np.random.random((5,3))
print(rand_arr)

# Định dạng in ra tối đa 3 chữ số phần thập phân

np.set_printoptions(precision=3)
print(rand_arr)

# Các phần tử có giá trị nhỏ hơn 1e-4 sẽ in dưới dạng mũ

rand_arr = rand_arr/1e3

print(rand_arr)

# Định dạng in ra dạng bình thường (không phải dạng 1e-x)

np.set_printoptions(suppress=True, precision= 6)
print(rand_arr)

# Để thiết lập lại định dạng in mặc định, dùng câu lệnh dưới đây
np.set_printoptions(edgeitems=3,infstr='inf',linewidth=75, nanstr='nan', precision=8,suppress=False, threshold=1000, formatter=None)


# In ra tối da 6 phần tử trong mảng

arr = np.arange(1,101)
np.set_printoptions(threshold=6)
print(arr)

# In ra tất cả các phần tử trong mảng
import sys
np.set_printoptions(threshold=sys.maxsize)
print(arr)
````

## Bài 19. Làm việc với dữ liệu từ file văn bản

* Nhập dữ liệu từ tập iris từ địa chỉ URL [iris dataset](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data)
https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data

* Lấy ra từng cột của dữ liệu

* Chuyển từ  mảng 1 chiều các Bộ (tuple) sang mảng 2 chiều numpy

* Tính trung bình, trung vị, độ lệch chuẩn

* Chuẩn hóa dữ liệu để các giá trị nằm trong khoảng từ 0 đến 1

````
# Bài 19
````

### Hướng dẫn bài 19


````
import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype='object')
names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')

print('Kích thước dữ liệu = ',iris.shape)

print('3 dòng đầu của dữ liệu\n',iris[:3])

print('Kiểu dữ liệu',type(iris[0,0]))
print(float(iris[0,0]))

# Nếu không đặt dtype = object, thì kết quả sẽ là 1 mảng các bộ
iris = np.genfromtxt(url, delimiter=',', dtype=None,encoding='utf-8')
print('Kích thước dữ liệu = ',iris.shape)

print('3 dòng đầu của dữ liệu\n',iris[:3])
print('Kiểu dữ liệu',type(iris[0][0]))

# Lấy ra 1 cột trong dữ liệu, ví dụ cột chứa các nhãn của dữ liệu

label = np.array([row[4] for row in iris])

print('5 nhãn đầu tiên: ',label[:5])

# Chuyển mảng 1 chiều các bộ (tuple) sang mảng 2 chiều numpy chứa dữ liệu
# Dữ liệu sẽ có kiểu là string

iris_2d = np.array([row.tolist()[:] for row in iris])
print('4 dòng đầu \n',iris_2d[:4])

# Nếu chỉ muốn lấy 4 cột đầu để tính toán (bỏ qua cột nhãn)
# Dữ liệu sẽ có kiểu là float
iris_2d = np.array([row.tolist()[:4] for row in iris])
print('4 dòng đầu \n',iris_2d[:4])

# Ta cũng có thể nhập trực tiếp từ file văn bản

iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
print('4 dòng đầu \n',iris_2d[:4])

# Tính giá trị trung bình, trung vị, độ lệch chuẩn của từng cột dữ liệu

print('Các giá trị thống kê')
for i in range(0,iris_2d.shape[1]):
  me,med,sd = np.mean(iris_2d[:,[i]]), np.median(iris_2d[:,[i]]), np.std(iris_2d[:,[i]])
  print(me,med,sd)

# Chuẩn hóa các từng cột dữ liệu, để các giá trị nằm trong đoạn [0, 1]

for i in range(0,iris_2d.shape[1]):
  imax, imin = iris_2d[:,[i]].max(), iris_2d[:,[i]].min()
  iris_2d[:,[i]] = (iris_2d[:,[i]] - imin)/(imax - imin)

print('5 dòng đầu của dữ liệu sau khi chuẩn hóa\n',iris_2d[:5])

````

````
# Bài 20
````

### Hướng dẫn bài 20


````
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
print('4 dòng đầu \n',iris_2d[:4])

def softmax(x):
  ex = np.exp(x)
  return ex/ex.sum()

for i in range(0, iris_2d.shape[1]):
  print('Giá trị theo softmax của 5 phần tử đầu\n',softmax(iris_2d[:,[i]])[:5])
````
