#Thư viện pandas là gì?

Thư viện pandas trong python là một thư viện mã nguồn mở, hỗ trợ đắc lực trong thao tác dữ liệu. Đây cũng là bộ công cụ phân tích và xử lý dữ liệu mạnh mẽ của ngôn ngữ lập trình python. Thư viện này được sử dụng rộng rãi trong cả nghiên cứu lẫn phát triển các ứng dụng về khoa học dữ liệu. Thư viện này có 2 câu trúc dữ liệu lõi là Series và DataFrame. Pandas cung cấp rất nhiều chức năng xử lý và làm việc trên cấu trúc dữ liệu này, làm cho việc xử lý dữ liệu có nhãn "labeled" hoặc dữ liệu quan hệ "relational" trở nên nhanh chóng, thuận tiện, dễ dàng và trực quan hơn. Chính sự linh hoạt và hiệu quả đã khiến cho pandas được sử dụng rộng rãi trong việc xử lý và phân tích dữ liệu.

#Tại sao sử dụng thư viện pandas?

* DataFrame đem lại sự linh hoạt và hiệu quả trong thao tác dữ liệu và lập chỉ mục;
* Là một công cụ cho phép đọc/ ghi dữ liệu giữa bộ nhớ với nhiều định dạng file: csv, text, excel, sql database, hdf5;
* Liên kết dữ liệu thông minh, xử lý được trường hợp dữ liệu bị thiếu. Tự động đưa dữ liệu phi cấu trúc về dạng có cấu trúc;
* Dễ dàng thay đổi bố cục của dữ liệu;
* Tích hợp cơ chế cắt (slice), lập chỉ mục, lấy ra tập con từ tập dữ liệu lớn.
* Có thể thêm, xóa các cột dữ liệu;
* Tập hợp hoặc thay đổi dữ liệu với group by cho phép bạn thực hiện các toán tử trên tập dữ liệu;
* Hiệu quả cao trong việc trộn và kết hợp các tập dữ liệu;
* Tối ưu về hiệu năng;
* Pandas được sử dụng rộng rãi trong cả học thuật và thương mại. Bao gồm thống kê, thương mại, phân tích, quảng cáo,…

# Series

Series là mảng một chiều giống như mảng Numpy, nhưng nó bao gồm thêm một bảng đánh label. Series có thể được khởi tạo thông qua NumPy, kiểu Dict hoặc các dữ liệu vô hướng bình thường

```
import calendar as cal
import numpy as np
import pandas as pd

# From NumPy
monthNames=[cal.month_name[i] for i in np.arange(1,6)]
print(monthNames)
months=pd.Series(np.arange(1,6),index=monthNames)
print(months)

# Lấy giá trị thông qua key
print(months.get('January'))

# Thêm giá trị

months.at['June'] = 6

print(months)

# From dictionary
currDict={'US' : 'dollar', 'UK' : 'pound',
          'Germany': 'euro', 'Mexico':'peso',
          'Nigeria':'naira',
           'China':'yuan', 'Japan':'yen'}

print(currDict)
currSeries=pd.Series(currDict)
print(currSeries)

# From scalar values

dogSeries=pd.Series('chihuahua', index=['breed','breed', 'countryOfOrigin', 'name', 'gender'])
print(dogSeries)

# From list, with default index

s = pd.Series([1,2,3,4,5,6,1,2,3,4,5,6])

print(s)

```

# Dataframe
![](https://storage.googleapis.com/lds-media/images/series-and-dataframe.width-1200.png)

DataFrame là mảng hai chiều được gán nhãn, nhưng không giống như NumPy, kiểu dữ liệu các cột là không đồng nhất, cấu trúc dữ liệu thì giống với NumPy nhưng có khả năng thay đổi. DataFrame gồm các thuộc tính như sau:

* Có khái niệm như một Table hoặc  SpreadSheet dữ liệu
* Tương tự với mảng NumPy nhưng không phải là subclass của np.ndarray
* Các cột dữ liệu là các kiểu không đồng nhất: float64, int, bool, …
* Các cột của DataFrame là cấu trúc của Series
* Nó có thể xem như một kiểu dữ liệu Dict của một cấu trúc Series nhưng cả hai * cột và hàng đều được đánh chỉ mục (index)
* Kích thước table có thể thay đổi: các cột có thể thêm hoặc xoá đi

Với mỗi chiều của Series/DataFrame được đánh chỉ mục, dù bất kể là mặc định
hay không. Các chỉ mục là cần thiết cho việc tim kiếm hoặc để kết nối dữ liệu phù hợp trong pandas. Các trục cần được đặt tên, ví dụ như mảng của các tháng “month” bao gôm các giá trị [Jan, Feb, … Dec]. Đây là cách DataFrame biểu diễn các chỉ mục, tên cột và tên các cột chỉ mục với các kí tự V, W, X, Y, Z:

* DataFrame được khởi tạo thông qua nhiều cách khác nhau như
* Dict của mảng một chiều, list, dict hoặc cấu trúc Series
*Mảng NumPy hai chiều
*Cấu trúc hoặc các bản ghi của ndarray
*Cấu trúc Series
*Các DataFrame khác
*Các nhãn của chỉ mục hay nhãn của cột có thể chỉ định theo dữ liệu, nếu không chỉ định thì sẽ được tự động sinh ra

```
import numpy as np
import pandas as pd
# DataFrame from Dict
stockSummaries={
'AMZN': pd.Series([346.15,0.59,459,0.52,589.8,158.88],
        index=['Closing price','EPS',
                'Shares Outstanding(M)',
                'Beta', 'P/E','Market Cap(B)']),
'GOOG': pd.Series([1133.43,36.05,335.83,0.87,31.44,380.64],
        index=['Closing price','EPS','Shares Outstanding(M)',
               'Beta','P/E','Market Cap(B)']),
'FB': pd.Series([61.48,0.59,2450,104.93,150.92],
      index=['Closing price','EPS','Shares Outstanding(M)',
             'P/E', 'Market Cap(B)']),
'YHOO': pd.Series([34.90,1.27,1010,27.48,0.66,35.36],
        index=['Closing price','EPS','Shares Outstanding(M)',
               'P/E','Beta', 'Market Cap(B)']),
'TWTR':pd.Series([65.25,-0.3,555.2,36.23],
       index=['Closing price','EPS','Shares Outstanding(M)',
              'Market Cap(B)']),
'AAPL':pd.Series([501.53,40.32,892.45,12.44,447.59,0.84],
       index=['Closing price','EPS','Shares Outstanding(M)','P/E',
              'Market Cap(B)','Beta'])}
stockDF=pd.DataFrame(stockSummaries)

print(stockDF)

stockDF = stockDF.transpose()
print(stockDF)
```

# Một số thao tác cơ bản trên Dataframe


Lấy giá trị một cột, nhiều cột, theo điều kiện


```
# Lấy giá trị một cột
print(stockDF['EPS'],'\n')

# Lấy giá trị nhiều cột

print(stockDF[['EPS','Closing price']],'\n')

# Lấy giá trị một hàng theo loc, iloc

## loc ưu tiên tìm theo nhãn label
print('loc\n',stockDF.loc['GOOG'],'\n')

#print(stockDF.loc[0,:],'\n')
## iloc ưu tiên tìm theo index ( chỉ số hàng)
print('iloc\n',stockDF.iloc[0],'\n')
print('iloc\n',stockDF.iloc[1,:3],'\n')

## Lấy hàng 2, 3 cột 2, 4

print('iloc [2,3][1,4\n',stockDF.iloc[[2,3],[1,4]])



## Lấy ra tên nhãn của 1 hàng
print('\nLabel of a row', stockDF.iloc[3]._name)
## Lấy ra các nhãn của cột
print("\nLabels of a column ",stockDF.iloc[3].index)

# Cũng có thể lấy ra giá trị một hàng như sau:
print(stockDF['GOOG':'GOOG'],'\n')


# Lấy ra các hàng có Beta > 0.7

print(stockDF[stockDF['Beta'] > 0.7],'\n')
```

Lấy một số hàng đầu, cuối của dữ liệu bằng hàm head, tail

```
print('3 top row\n',stockDF.head(3))

print('3 last row\n',stockDF.tail(3))
```

Thêm cột mới bằng cách gán


```
stockDF['VIN'] = 2019
stockDF['IDX'] = [i for i in range(len(stockDF))]
print(stockDF)
```

Xóa cột


```
stockDF = stockDF.drop(columns = ['VIN','IDX'] )
print(stockDF)
```

Xóa hàng

```
# Xóa hàng theo nhãn mỗi hàng
print( stockDF.drop(['GOOG','YHOO']),'\n')

# Xóa hàng theo chỉ số

print( stockDF.drop([stockDF.index[0],stockDF.index[1]]))

```

Chú ý các lệnh xóa trên không làm thay đổi dataFrame, nó chỉ trả lại datFrame mới đã xóa các phần tử theo yêu cầu, để thực sự xóa dữ liệu trong dataFrame chúng ta đặt tham số inplace = True

```
## Xóa 2 hàng
print( stockDF.drop(['GOOG','YHOO']),'\n')

## Dữ liệu ban đầu không đổi
print(stockDF)

## Xóa thực 2 hàng trong dữ liệu ban đầu (inplace = True)
stockDF.drop(['GOOG','YHOO'], inplace=True)
## Dữ liệu ban đầu đã bị xóa
print('Dữ liệu sau khi bị xóa\n',stockDF,'\n')
```
