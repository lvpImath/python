## Hướng dẫn bài 1

````
# Inputs
import numpy as np
import pandas as pd
mylist = list('abcedfghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

print(mylist)
print(myarr)
print(mydict)

# Solution
ser1 = pd.Series(mylist)
ser2 = pd.Series(myarr)
ser3 = pd.Series(mydict)

#print(ser1,ser2,ser3,sep='\n')

dfser3 = ser3.to_frame().reset_index()


df = pd.DataFrame({'col1':ser1,'col2':ser2,'col3':dfser3['index']})
print(df.head())
````
